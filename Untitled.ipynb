```python
X = df.drop(['attack_detected'], axis=1)
y = df['attack_detected']
```

```python
df.head()
```

```python
numerical_features = ['network_packet_size', 'login_attempts', 'session_duration', 'ip_reputation_score', 'failed_logins', 'unusual_time_access']
categorical_features = ['protocol_type', 'encryption_used', 'browser_type']
```

```python
all_features = numerical_features + categorical_features
missing_features = [col for col in all_features if col not in X.columns]

if missing_features:
    print(f"\nFATAL ERROR: The following required features are missing from the data: {missing_features}")
    print("Please check for typos in the feature lists or verify the column names in the CSV file.")
    exit()
else:
    print("\nAll required features found in the DataFrame. Proceeding with preprocessing.")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
```

```python
numerical_transformer = Pipeline(steps=[
    ('scaler', StandardScaler())
])
```

```python
categorical_transformer = Pipeline(steps=[
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, numerical_features),
        ('cat', categorical_transformer, categorical_features)
    ],
    remainder='passthrough'
)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
print(f"\nTraining set size: {X_train.shape[0]} samples")
print(f"Testing set size: {X_test.shape[0]} samples")
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
models = {
    "Logistic Regression (for Binary Classification)": LogisticRegression(),
    "Decision Tree Classifier": DecisionTreeClassifier(),
    "Support Vector Classifier (SVC)": SVC(),
    "Random Forest Classifier": RandomForestClassifier(),
    "K-Nearest Neighbors (KNN)": KNeighborsClassifier(),
}
```

```python
results = {}
```

```python
for name, model in models.items():
    
    pipeline = Pipeline(steps=[
        ('preprocessor', preprocessor),
        ('classifier', model)
    ])
    print(f"Training {name}...")
    pipeline.fit(X_train, y_train)
    y_pred = pipeline.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    results[name] = accuracy

    print(f"  -> Accuracy: {accuracy:.4f}")
best_model_name = max(results, key=results.get)
best_accuracy = results[best_model_name]

print("\n--- Summary of Results ---")
for name, acc in results.items():
    print(f"| {name.ljust(43)} | Accuracy: {acc:.4f} |")

print("\n--- Best Performing Algorithm ---")
print(f"The best model is: {best_model_name}")
print(f"With the highest accuracy of: {best_accuracy:.4f}")
```

```python
print(f"\nCreating and training the final pipeline for {best_model_name} on the full training set...")
final_model = RandomForestClassifier(random_state=42)
final_model = models[best_model_name]
final_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', final_model)
])
final_pipeline.fit(X_train, y_train)
print("Final model training complete.")
```

```python
def predict_new_sample(pipeline, sample_data):
    """
    Predicts 'attack_detected' for a new, single sample.

    Args:
        pipeline: The trained scikit-learn Pipeline (final_pipeline).
        sample_data: A dictionary containing the features of the new sample.
                     Keys must match the column names used in training.

    Returns:
        The predicted class (0 for no attack, 1 for attack detected).
    """

    sample_df = pd.DataFrame([sample_data])

    prediction = pipeline.predict(sample_df)[0]
    

    result_text = "Attack Detected (1)" if prediction == 1 else "No Attack Detected (0)"
    print(f"\n--- Prediction Result ---")
    print(f"Input Data:\n{sample_data}")
    print(f"Model Prediction: {result_text}")
    return prediction
```
